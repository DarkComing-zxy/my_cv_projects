# 1.数据集准备

代码所需要的包

```python
from tensorflow import keras
from tensorflow.keras import layers
import os
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

下载cifar-10数据集

```python
(train_datas,train_labels),(test_datas,test_labels) = keras.datasets.cifar10.load_data()
```

cifar-10数据集标签：



```python
print(train_datas.shape,train_labels.shape)
```

划分数据集

```python
dataset = tf.data.Dataset.from_tensor_slices((train_datas,train_labels))
dataset = dataset.shuffle(buffer_size = len(train_datas))
train_dataset = dataset.take(int(0.8 * len(train_datas))).batch(32)
val_dataset = dataset.skip(int(0.8 * len(train_datas))).batch(32)
test_dataset = tf.data.Dataset.from_tensor_slices((test_datas,test_labels)).batch(32)
print(f"训练集样本数:{len(train_dataset)}")
print(f"验证集样本数:{len(val_dataset)}")
```

# 2.构建模型并训练

## 2.1-下载并微调预训练模型vgg16

```python
conv_base = keras.applications.vgg16.VGG16(include_top = False,input_shape = (32,32,3))
```

```python
conv_base.summary()
```

```python
for layer in conv_base.layers[:-4]:
    layer.trainable = False
```

### 2.1.1-基础模型

```python
inputs = keras.Input(shape = (32,32,3))
x = keras.applications.vgg16.preprocess_input(inputs)
x = conv_base(x)
x = layers.Flatten()(x)
x = layers.Dense(256,activation = "relu")(x)
outputs = layers.Dense(10,activation = "softmax")(x)
model = keras.Model(inputs,outputs)
model.summary()
```

```python
model.compile(
    optimizer = "rmsprop",
    loss = "sparse_categorical_crossentropy",
    metrics = ["accuracy"]
)
callbacks = [
    keras.callbacks.ModelCheckpoint(
        filepath = "Base_CNN.keras",
        monitor = "val_loss",
        save_best_only = True
    )
]
history = model.fit(
    train_dataset,
    epochs = 50,
    validation_data = val_dataset,
    callbacks = callbacks
)
```

```python
history.history.keys()
```

绘制精度和损失曲线

```python
epochs = range(1,len(history.history["loss"]) + 1)
train_acc = history.history["accuracy"]
train_loss = history.history["loss"]
val_loss = history.history["val_loss"]
val_acc = history.history["val_accuracy"]
plt.figure(figsize = (12,5))
plt.subplot(1,2,1)
plt.plot(epochs,train_acc,"b",label = "Training Accuracy")
plt.plot(epochs,val_acc,"ro",label = "Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Training and validation accuracy")
plt.subplot(1,2,2)
plt.plot(epochs,train_loss,"b",label = "Training Loss")
plt.plot(epochs,val_loss,"ro",label = "Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training and validation loss")
plt.legend()
plt.tight_layout()
plt.show()
```

在测试集上运行

```python
model = keras.models.load_model("Base_CNN.keras")
result = model.evaluate(test_dataset)
```

观察精度曲线和损失曲线可以发现一个明显的现象：
1. 验证集精度逐渐拟合，曲线平滑
2. 验证集损失先下降，然后发散不拟合，曲线抖动幅度很大
现象解释：

|置信度|精度|交叉熵|
|----|----|------|
|0.99|1|0.01|
|0.80|1|0.22|
|0.60|1|0.51|
|0.40(预测错误)|0|0.92|

随着训练轮数的增加，模型出现过拟合，虽然对验证集的预测是对的，即精度为1，但是置信度下降，交叉熵变大，最终的表现就是精度拟合，损失发散  
结论：
- 精度只看对不对，损失还要看自不自信
- 损失对过拟合更加敏感，这也是为什么监控要用损失而不是精度
- 验证集损失先降后升而精度仍涨 → 典型的过拟合信号

### 2.1.4-使用dropout

```python
data_augmentation = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.2)
])
inputs = keras.Input(shape = (32,32,3))
x = data_augmentation(inputs)
x = keras.applications.vgg16.preprocess_input(x)
x = conv_base(x)
x = layers.Flatten()(x)
x = layers.Dropout(0.5)(x)
x = layers.Dense(256,activation = "relu")(x)
outputs = layers.Dense(10,activation = "softmax")(x)
model = keras.Model(inputs,outputs)
model.summary()
```

```python
model.compile(
    optimizer = keras.optimizers.RMSprop(learning_rate = 1e-4),
    loss = "sparse_categorical_crossentropy",
    metrics = ["accuracy"]
)
callbacks = [
    keras.callbacks.ModelCheckpoint(
        filepath = "Dropout.keras",
        monitor = "val_loss",
        save_best_only = True
    )
]
history = model.fit(
    train_dataset,
    epochs = 50,
    validation_data = val_dataset,
    callbacks = callbacks
)
epochs = range(1,len(history.history["loss"]) + 1)
train_acc = history.history["accuracy"]
train_loss = history.history["loss"]
val_loss = history.history["val_loss"]
val_acc = history.history["val_accuracy"]
plt.figure(figsize = (12,5))
plt.subplot(1,2,1)
plt.plot(epochs,train_acc,"b",label = "Training Accuracy")
plt.plot(epochs,val_acc,"ro",label = "Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Training and validation accuracy")
plt.subplot(1,2,2)
plt.plot(epochs,train_loss,"b",label = "Training Loss")
plt.plot(epochs,val_loss,"ro",label = "Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training and validation loss")
plt.legend()
plt.tight_layout()
plt.show()
```

### 2.1.5-图像resize

```python
def preprocess_image(image,label):
    image = tf.image.resize(image,(224,224))
    return image,label
train_dataset = train_dataset.map(preprocess_image,num_parallel_calls = tf.data.AUTOTUNE).prefetch(tf.data.AUTOTUNE)
val_dataset = val_dataset.map(preprocess_image,num_parallel_calls = tf.data.AUTOTUNE).prefetch(tf.data.AUTOTUNE)
test_dataset = test_dataset.map(preprocess_image,num_parallel_calls = tf.data.AUTOTUNE).prefetch(tf.data.AUTOTUNE)
```

```python
data_augmentation = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.2)
])
conv_base = keras.applications.vgg16.VGG16(include_top = False,weights = "imagenet",input_shape = (224,224,3))
for layer in conv_base.layers[:-4]:
    layer.trainable = False
for layer in conv_base.layers[-4:]:
    layer.trainable = True
inputs = keras.Input(shape = (224,224,3))
x = data_augmentation(inputs)
x = keras.applications.vgg16.preprocess_input(x)
x = conv_base(x)
x = layers.Flatten()(x)
x = layers.Dropout(0.5)(x)
x = layers.Dense(256,activation = "relu")(x)
outputs = layers.Dense(10,activation = "softmax")(x)
model = keras.Model(inputs,outputs)
model.summary()
```

```python
model.compile(
    optimizer = keras.optimizers.RMSprop(learning_rate = 1e-4),
    loss = "sparse_categorical_crossentropy",
    metrics = ["accuracy"]
)
callbacks = [
    keras.callbacks.ModelCheckpoint(
        filepath = "Dropout.keras",
        monitor = "val_loss",
        save_best_only = True
    ),
    keras.callbacks.ReduceLROnPlateau(
        factor = 0.5,
        min_lr = 1e-7,
        patience = 3
    )
]
history = model.fit(
    train_dataset,
    epochs = 30,
    validation_data = val_dataset,
    callbacks = callbacks
)
epochs = range(1,len(history.history["loss"]) + 1)
train_acc = history.history["accuracy"]
train_loss = history.history["loss"]
val_loss = history.history["val_loss"]
val_acc = history.history["val_accuracy"]
plt.figure(figsize = (12,5))
plt.subplot(1,2,1)
plt.plot(epochs,train_acc,"b",label = "Training Accuracy")
plt.plot(epochs,val_acc,"ro",label = "Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Training and validation accuracy")
plt.subplot(1,2,2)
plt.plot(epochs,train_loss,"b",label = "Training Loss")
plt.plot(epochs,val_loss,"ro",label = "Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training and validation loss")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
cat_dog_datas = []
cat_dog_labels = []
x_train = []
y_train = []
train_dataset = train_dataset
for image,label in train_dataset:
    x_train.append(image)
    y_train.append(label)
print(len(x_train),len(y_train))
```

```python
for image,label in zip(x_train,y_train):
    if int(label) == 3 or int(label) == 5:
        cat_dog_datas.append(image)
        cat_dog_labels.append(label)
print(len(cat_dog_datas),len(cat_dog_labels))
```

```python
augmentation = keras.Sequential([
    layers.RandomFlip("horizontal"),         # 随机水平翻转
    layers.RandomRotation(0.2),             # 随机旋转 ±20%
    layers.RandomZoom(0.1),                 # 随机缩放 ±10%
    layers.RandomContrast(0.2)              # 随机对比度 ±20%
])
balanced_train_dataset = (
    train_dataset
    .filter(lambda x,y: tf.reduce_any(tf.equal(y,[3,5])))
    .map(lambda x,y: (augmentation(tf.expand_dims(x,0), training=True)[0], y))
    .batch(32)
    .shuffle(60000)
    .repeat()  # 无限重复
    .prefetch(tf.data.AUTOTUNE)
)
```

```python
model = keras.models.load_model("Train_More.keras")
callbacks = [
    keras.callbacks.ModelCheckpoint(
        filepath = "Finnal.keras",
        monitor = "val_loss",
        save_best_only = True
    ),
    keras.callbacks.ReduceLROnPlateau(
        factor = 0.5,
        min_lr = 1e-7,
        patience = 3
    )
]
history = model.fit(
    balanced_train_dataset,
    validation_data=val_dataset,
    epochs=30,  # 例如之前训练了50轮，这里写80
    callbacks = callbacks,
    steps_per_epoch = 1500
)
epochs = range(1,len(history.history["loss"]) + 1)
train_acc = history.history["accuracy"]
train_loss = history.history["loss"]
val_loss = history.history["val_loss"]
val_acc = history.history["val_accuracy"]
plt.figure(figsize = (12,5))
plt.subplot(1,2,1)
plt.plot(epochs,train_acc,"b",label = "Training Accuracy")
plt.plot(epochs,val_acc,"ro",label = "Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Training and validation accuracy")
plt.subplot(1,2,2)
plt.plot(epochs,train_loss,"b",label = "Training Loss")
plt.plot(epochs,val_loss,"ro",label = "Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training and validation loss")
plt.legend()
plt.tight_layout()
plt.show()
```
